**3. Escolha uma coluna com valores numÃ©ricos e apague todas as linhas que possuam valor nulo**


```python
dados_novo= dados.dropna(subset=['Trestbps'])
display(dados_novo['Trestbps'])
```

**4. Calcule a mÃ©dia desta mesma coluna**

```python
media = dados['Age'].mean()
print(media)
```

**5. Escolha um critÃ©rio para eliminar as linhas que possuam valores muito distantes da mÃ©dia e crie um novo Dataframe.**

```python
# CritÃ©rio escolhido: o valor tem que estar a um mÃ¡ximo de (+ / - 10) da mÃ©dia

criterio1 = dados['Age'] < media+10
novo_dados = dados[criterio1]
criterio2 = novo_dados['Age'] > media-10
novo_dados = novo_dados[criterio2]
display(novo_dados['Age'])
```

**6. Crie um novo Dataframe que contenha apenas 5 colunas do conjunto anterior.**

```python
dados_cinco_colunas = novo_dados[['Age','Sex','Oldpeak', 'Ca', 'Thal']]
display(dados_cinco_colunas)
```

**7. Escolha duas colunas numÃ©ricas e crie uma nova coluna que siga a regra: (Coluna A - Coluna B)*nÃºmero aleatÃ³rio (entre 0 e 1).**

```python
#Coluna A = Trestbps   // Coluna B = Slope
import random

dados.insert(1,'coluna nova',(dados['Trestbps']-dados['Slope'])*random.randint(0,1))

display(dados)
```

**8. Embaralhe aleatoriamente o dataframe (Ã­ndices e valores).**

```python
import numpy as np

dados = dados.reindex(np.random.permutation(dados.index))

display(dados)
```

**9. Ordene o Dataframe seguindo um critÃ©rio definido por vocÃª (Explique o critÃ©rio)**

```python
#CritÃ©rio: Ordenar de modo CRESCENTE a partir da coluna escolhida --> "Chol" 

dados = dados.sort_values(by='Chol', ascending=True)
display(dados)
```

**10. Escolha uma coluna numÃ©rica e um coluna com strings e utilize a funÃ§Ã£o Groupby para mostrar os totais da  coluna.**

```python
#Coluna numÃ©rica escolhida = coluna "CP"
#Coluna com strings escolhida = coluna "Ca"

grupo = dados.groupby('CP')[['Ca']].count()
print(grupo)
```

**11. Utilize a funÃ§Ã£o MAPS para fazer algum tipo de substituiÃ§Ã£o no Dataframe (explique)**

```python
#A partir da leitura da coluna "Sex" Ã© identificado e adicionado a partir do Maps a coluna "Meaning_Sex" que escreve se a pessoa Ã© "male"(sexo masculino) ou "female"(sexo feminino)
dados['Meaning_Sex'] = dados['Sex'].map(
    {0: 'female', 1: 'male'})

display(dados)
```
